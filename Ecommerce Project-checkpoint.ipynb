```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x, startangle=60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Ship Mode",data=dataset,hue="Category")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Segment",data=dataset)
plt.figtext(0.5, -0.1, 
            "Most orders come from the Consumer segment, followed by Corporate and Home Office segments.", 
            wrap=True, ha="center", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Category", data=dataset, palette="Set2")
plt.title("Orders by Category")

# Short conclusion
plt.figtext(0.5, -0.1, 
            "Office Supplies has the highest number of orders, followed by Technology and Furniture.", 
            wrap=True, ha="center", fontsize=10)

plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="Category",data=dataset[dataset['Category']=='Furniture'],hue="Sub-Category")
plt.figtext(0.5, -0.15, 
            "Among Furniture items, some sub-categories like Chairs and Tables dominate the sales.", 
            wrap=True, ha="center", fontsize=10)
plt.show()
```

```python
print(dataset["Order Date"].head())
print(dataset["Order Date"].dtype)
```

```python
dataset["Order Date"] = pd.to_datetime(dataset["Order Date"], errors="coerce", dayfirst=True)
```

```python
dataset["Year"] = dataset["Order Date"].dt.year
print(dataset[["Order Date", "Year"]].head())
```

```python
dataset.info()
```

```python
dataset["Year"].value_counts()
```

```python
plt.figure(figsize=(8,4))
plt.hist(dataset["Year"], bins=10, edgecolor="Red")
plt.title("Year")
plt.figtext(0.5, -0.1, 
            "Most orders are concentrated around recent years, showing growth over time.", 
            wrap=True, ha="center", fontsize=10)
plt.show()
```

```python
sns.barplot(x="Category", y="Profit",data=dataset,estimator="sum")
plt.figtext(0.5, -0.1, 
            "Office Supplies generate the highest total profit, followed by Technology and Furniture.", 
            wrap=True, ha="center", fontsize=10)
plt.show
```

```python
dataset["State"].value_counts()[:5]
```

```python

top_10_state = dataset["State"].value_counts().head(10)
plt.figure(figsize=(12,5))
sns.countplot(
    data=dataset[dataset["State"].isin(top_10_state.index)],
    x="State",
    order=top_10_state.index,
    hue="State",   
    palette="tab10",
    legend=False
)
plt.figtext(0.5, -0.1, 
            "California has the highest orders, followed by Texas and New York, showing strong sales in these states compared to others in the top 10", 
            wrap=True, ha="center", fontsize=10)

plt.show()
```
